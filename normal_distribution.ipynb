```python
# Z- score : How many std deviation away from a datapoint is from mean
```

```python
df.head()
```

```python
df['zscore']= (df.height - df.height.mean())/df.height.std()
df.head()
```

```python
df[(df.zscore<-3 ) | (df.zscore>3 )]
```

```python
df_no_outlier = df[(df.zscore<-3 ) | (df.zscore>3 )]
df_no_outlier
```
